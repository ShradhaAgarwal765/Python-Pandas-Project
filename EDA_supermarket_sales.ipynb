```python
days_map = {
    0:'Monday',
    1:'Tuesday',
    2:'Wednesday',
    3:'Thursday',
    4:'Friday',
    5:'Saturday',
    6:'Sunday',
}
```

```python
df1['dayofweek'] = df1['Date'].dt.dayofweek.map(days_map)
df1['dayofweek']
```

```python
df1['dayofweek'].value_counts().plot(kind='bar')
```

```python
map_month = {
    1:'Jan',
    2:'Feb',
    3:'Mar',
}

df1['month'] = df1['Date'].dt.month.map(map_month)
df1['month']
```

```python
df1.month.value_counts().plot(kind='bar')
```
